```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel("2019 Top MLB Batters.xlsx")
```

```python
plt.figure()
plt.plot(df['Homeruns'],df['BA'],'*',color="purple", label='Comparing Homeruns to Batting Average ')
plt.xlabel('Homeruns')
plt.ylabel('BA')
plt.legend()
```
